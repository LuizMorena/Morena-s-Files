```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Carregando o Dataset 

```python
atletas = pd.read_csv('athlete_events.csv')
regiao = pd.read_csv('noc_regions.csv')
regiao.head(50)
```

### Unindo os Datasets 

```python
atletas_df = atletas.merge(regiao, how = 'left', on = 'NOC')
atletas_df.head(50)
```

```python
atletas_df.shape
```

### Corrigindo nome das colunas 

```python
atletas_df.rename(columns = {'Name':'Nome','Sex':'Sexo', 'Age':'Idade', 'Height':'Altura', 'Weight':'Peso', 
                              'Team':'Confederao', 'Games':'Jogos', 'Year':'Ano', 'Season':'Estao', 'City':'Cidade', 
                             'Sport':'Esporte', 'Event':'Evento', 'Medal':'Medalha', 'region':'Regio', 'notes':'Notas'},
                 inplace=True)
```

```python
atletas_df.head()
```

```python
atletas_df.info()
```

```python
atletas_df.describe()
```

### Checando Valores Nulos

```python
nan_valores = atletas_df.isna()
nan_colunas = nan_valores.any()
nan_colunas
```

###  Top 10 Pases participantes

```python
top_10_paises = atletas_df.Confederao.value_counts().sort_values(ascending=False).head(10)
top_10_paises
```

### Grfico para Top 10 Pases 

```python
plt.figure(figsize=(12,6))
#plt.xticks(rotation=20)
plt.title('Participao Geral de Pases')
sns.barplot(x=top_10_paises.index, y=top_10_paises, palette = 'Set2')
```

###  Distribuio de Idade por Atleta

### Atletas com Medalhas de Ouro 

```python
ouromed = atletas_df[(atletas_df.Medalha == 'Gold')]
ouromed.head(10)
```

### Mostrando somente os valores diferentes de NaN 

```python
ouromed = ouromed[np.isfinite(ouromed['Idade'])]
```

### Atletas Ouro com Mais de 60 Anos 

```python
ouromed['ID'][ouromed['Idade']>60].count()
```

### Em Quais Modalidades Vieram Estas Medalhas

```python
modalidade = ouromed['Esporte'][ouromed['Idade']>60]
modalidade
```

### Grfico de Modalidade 

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(modalidade)
plt.title('Medalhas de Ouro para Maiores de 60')
```

### Medalhas de Ouro por Pas

```python
ouromed.Regio.value_counts().reset_index(name='Medalha').head(5)
```

### Grfico de Medalhas de Ouro por Pas

```python
paisouromed = ouromed.Regio.value_counts().reset_index(name='Medalha').head(5)
g = sns.catplot(x='index', y='Medalha', data=paisouromed, height=5, kind='bar',palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 Pases')
g.set_ylabels('Nmero de Medalhas')
plt.title('Medalhas de Ouro por Pas')
```
