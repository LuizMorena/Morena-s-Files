```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# Checando se há dados faltando

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
    
```

```python
df = df.dropna()
```

```python
# Tipo de datas nas colunas

df.dtypes
```

```python
#Mudando o tipo dos dados das colunas 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#Criando a correta coluna de Ano
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 100)
```

```python
#Removendo Valores Duplicados

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
df
```

```python
df
```

```python
#Gráfico de Dispersão Custo x Bilheteria

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Custo x Bilheteria')

plt.xlabel('Bilheteria Bruta')
plt.ylabel('Custo por Filme')

plt.show
```

```python
df.head()
```

```python
# Custo x Bilheteria usando o seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red",}, line_kws={"color":"blue"})
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Matriz de Correlação por Caractéristicas Numéricas')

plt.xlabel('Carac dos Filmes')
plt.ylabel('Carac dos Filmes')

plt.show
```

```python
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'): 
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized        
        
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Matriz de Correlação por Caractéristicas Numéricas')

plt.xlabel('Carac dos Filmes')
plt.ylabel('Carac dos Filmes')

plt.show
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
